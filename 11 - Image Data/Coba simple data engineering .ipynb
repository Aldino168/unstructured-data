````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
plt.imread("data//0/1.png")
````

# X = Path ->Image->flatten

````python
X=[]
for i in file_path:
    X.append(np.array([plt.imread(i).flatten()]))
````

# y = Path -> split -> ambil index ke 1

````python
y= []
for j in file_path:
    y.append(np.array([int(j.split("\\")[1])]))
````
