```python
from glob import glob
```

```python
fname=glob('data/*/*.png')
fname[:10]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# X = Path -> Read Image (Matplotlib) -> Flatten

```python
for i in fname:
    np.array([plt.imread(i).flatten])
```

```python
X=np.array([plt.imread(i).flatten() for i in fname])
X
```
