```python
import re

def casefolding(text):
    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'[-+]?[0-9]+', '', text)
    text = re.sub(r'[^\w\s]','', text)
    text = re.sub(r'[@,.?[\]#]', '', text)
    text.strip()
    return text
```

```python
normalisasi = pd.read_csv('data/key_norm.csv')
def normalisasi_teks(text):
    text = ' '.join([normalisasi[normalisasi['singkat'] == word]['hasil'].values[0] if (normalisasi['singkat'] ==word).any() else word for word in text.split()])
    text = str.lower(text)
    return text
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

stopwords_ind = stopwords.words('indonesian')

def stop_words(text):
    clean_words=[]
    for word in text.split():
        if word not in stopwords_ind:
            clean_words.append(word)
    return " ".join(clean_words)
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(text):
    text = stemmer.stem(text)
    return text
```

```python
def preprocessing_text(text):
    text = casefolding(text)
    text = normalisasi_teks(text)
    text = stop_words(text)
    text = stemming(text)
    return text
```

```python
%%time
df_copy_baru['teks_setelah_prep'] = df_copy_baru['reviewContent'].apply(preprocessing_text)
```

**step step yang dilakukan:**

    > data acquisition
    
    > memahami data
    
    > preprocessing data (memilih kolom, menghapus data missing value)
    
    > preprocessing text (merubah token menjadi huruf kecil semua, casefolding, normalisasi data(slang kata), menghapus data yang kurang penting (stopwords), menghapus prefix dan sufix)
    
    > ekstraksi feature text menggunakan tfidf
    
    > seleksi fitur
    
    > melakukan preprocessing sedikit lebih lanjut 
    
    > melihat fitur penting dari tiap kategori target
    
**hal hal yang penting:**

* data yang digunakan adalah data review customer lazada, dengan berbagai fitur nya.
* kolom yang digunakan rating dan reviewContent
* terdapat missing value pada data reviewContent, oleh karena itu perlu manipulasi drop data pada baris yang missing value itu
* melakukan casefolding atau menghapus beberapa regular expression
* melakukan normalisasi token yang dianggap slang dengan data key_norm
* menghapus data yang dianggap kurang penting (biasanya dari kata-kata yang sering muncul, ex: di,itu,ini,dll
* merubah tiap token yang ada imbuhan nya menjadi bentuk dasar, dengan anggapan memiliki makna yang mirip (stemming)
* waktu yang diperlukan dalam preprocessing data teks dengan step step diatas sekitar 1 jam 13 menit.
* lalu dilakukan ekstraksi fitur menjadi vektor dari data yang telah di ekstraksi tersebut menggunakan tfidf, karena tfidf lebih sedikit baik karena memperhatikan dari dokumen lain untuk ekstraksi fitur nya. dan secara tidak langsung melemahkan fitur yang sering muncul di dokumen lain, karena terdapat step inverse document di tfidf ini.
* seleksi fitur menggunakan chi square dengan filtering.
* mencoba preprocessing data lebih lanjut, dengan harapan bisa improve makna guna dimasukkan ke dalam model machine learning kelak.
* jika dianalisis dari word cloud, fitur fitur yang ada pada tiap kategori sudah lumayan cukup merepresentasikan tiap kategori nya.
* menggunakan n_gram yang berbeda mengakibatkan jumlah ekstraksi fitur akan berbeda, jika menggunakan 1 dan 2 kata berurutan maka akan menyebabkan bertambahnya fitur yang diekstraksi. dalam penentuan n_gram ini sesuai dengan kebutuhan kasus yang ingin diselesaikan.
